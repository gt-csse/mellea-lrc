# Test helpers/convert_to_label_studio_format.py

```python
from typing import cast

from eyecite import get_citations, resolve_citations
from eyecite.models import Resource

from mellea_lrc.helpers import create_bibliography, send_data
```

## Some tests

```python
example_citation = "Methodist Hosp of Sacramento v. Shalala , 38 F.3d 1225, 1227 -28 (D.C. Cir. 1994)."
supra = "Methodist Hosp of Sacramento v. Shalala, supra."
short_citation = "38 F.3d, at 1225      ."
second_citation = "Samuels v. Mackell, 401 U. S. 66 (1971)."
s_id = "Id., at 66"
```

### Get citations

```python
new_lines = "\n" * 3
text = f"{example_citation}{new_lines}{short_citation}{new_lines}{second_citation}{new_lines}{s_id}{new_lines}{supra}"
citations = get_citations(text)
```

```python
resources = resolve_citations(citations)
```

```python
resources = cast(dict[Resource, list], resources)
biblio = create_bibliography(citations, resources)
```

```python
print(biblio)  # noqa: T201
```

```python
send_data(text, biblio)
```
