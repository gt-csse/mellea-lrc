```python
from typing import cast


from eyecite import get_citations, resolve_citations
from eyecite.models import Resource, CitationBase
```

```python
example_citation = "Methodist Hosp of Sacramento v. Shalala , 38 F.3d 1225, 1227 -28 (D.C. Cir. 1994)."
supra = "Methodist Hosp of Sacramento v. Shalala, supra."
short_citation = "38 F.3d, at 1225"
second_citation = "Samuels v. Mackell, 401 U. S. 66 (1971)."
s_id = "Id., at 66"
```

```python
citation = get_citations("\n".join([example_citation, short_citation, second_citation, s_id, supra]))
```

```python
for cite in citation:
    print(cite)
```

```python
resolved_citations = resolve_citations(citation)
```

```python
print(resolved_citations)
```

```python
rs = cast(dict[Resource, list[CitationBase]], resolved_citations)
```

```python
for key, item in rs.items():
    print(f"Key: {key} ")
    for cite in item:
        cite_type = type(cite).__name__
        from eyecite.models import FullCaseCitation

        if type(cite) is FullCaseCitation:
            print(f"     Value({cite_type}): {cite.metadata.plaintiff}")
        else:
            print(f"     Value({cite_type}): {cite}")

    print()
```
